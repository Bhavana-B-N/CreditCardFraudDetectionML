```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a Pandas Dataframe
credit_card_data = pd.read_csv("creditcard.csv")
```

```python
credit_card_data.tail(10)
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking the number of missing values in each column

credit_card_data.isnull().sum()
```

```python
#number of legit and fraudulent transactions
credit_card_data['Class'].value_counts()

#data is highly unbalanced
#0->legit
#1->fraud
```

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
fraud.head(10)
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transaction
#mean value is different for both the transaction
credit_card_data.groupby('Class').mean()
```

```python
#Under sampling the legit data
#to balance the data
legit_sample = legit.sample(n=492)
```

```python
# concatenating two dataframes legit sample,fraud
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.tail()
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
#mean values are almost same So there is no change in the distrubution of dataset
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X.info())
print(Y)
```

```python
#Splitting the data into Training data and Testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X_train.shape,X_test.shape)
```

```python
#Model training
#Logistic regression

model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
#Model Evaluation
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Training Accuracy score:",training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Training Accuracy score:",test_data_accuracy)
```
